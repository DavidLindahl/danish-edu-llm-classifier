```python
import pandas as pd

import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('danish_pretranslate_unbalanced.csv')

# Create histogram of int_score
plt.figure(figsize=(10, 6))
plt.hist(df['int_score'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of int_score')
plt.xlabel('int_score')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```
